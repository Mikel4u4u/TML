```python
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures

import matplotlib.pyplot as plt

%matplotlib inline
```

```python
X = np.linspace(-5, 5, num=100)[:, None]
y = -0.5 + 2.2*X +0.3*X**3+ 2*np.random.randn(100, 1)
```

```python
plt.plot(X, y)
```

```python
X.shape
model = LinearRegression()
poly = PolynomialFeatures(degree=3, include_bias=False)
X_new = poly.fit_transform(X)
X_new.shape
model.fit(X_new[:85], y[:85])
y_pred = model.predict(X_new[5:])

model.fit(X[:85], y[:85])
y_pred1 = model.predict(X[5:])

fig = plt.subplots( figsize=(10, 8))
plt.scatter(X, y)
plt.plot(X[5:], y_pred, 'r',);
plt.plot(X[5:], y_pred1, 'g',);
plt.legend(('Нелинейная регр','Линейная регр','Обучающая выборка'), loc='upper left')
```
